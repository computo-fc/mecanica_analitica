## Y, ¿ecuaciones de segundo orden?

Queremos resolver ahora
$$\ddot{x} + \omega^2 x = 0$$
utilizando la misma función `odeint`. ¿Será posible?

Sí: reducimos la ecuación a un sistema de ecuaciones de 1er. orden:
$$\dot{\mathbf{x}} = \mathbf{f}(\mathbf{x}) $$

Entonces necesitamos definir una función que acepte un vector y regrese otro vector:

```python
omega_cuad = 4

def armonico(x_vec, t):
    x, v = x_vec  # desmenuza el vector en sus 2 componentes
    
    f1 = v
    f2 = -omega_cuad * x
    
    return f1, f2
    
```

```python
1, 2
```

Esto se llama una *tupla*, o un par ordenado

```python
x_0 = 1
v_0 = 0

xvec_inic = x_0, v_0
```

```python
t = np.arange(0, 10, 0.01)
soln_armonica = odeint(armonico, xvec_inic, t) 
```

```python
soln_armonica
```

```python
soln_armonica[0][1]  # contamos desde 0 siempre
```

```python
soln_armonica[0, 1]
```

```python
soln_armonica[0:5, 1]
```

```python
x = soln_armonica[:, 0]  # todos
v = soln_armonica[:, 1]
```

```python
plt.plot(t, x)
plt.plot(t, v)
```

```python
plt.plot(x, v)
plt.axis("scaled")
plt.plot([x_0], [v_0], 'o', markersize=10)
plt.xlim(-1.2, 1.2)
plt.ylim(-2.2, 2.2)
```
