## Hacia las ecuaciones diferenciales ordinarias

Queremos resolver
$$\dot{\mathbf{x}} = \mathbf{f}(\mathbf{x}, t)$$
sujeta a 
$$\mathbf{x}(t=t_0) = \mathbf{x}_0$$

Pero no se puede (analíticamente).

Numéricamente, *todo* se puede [excepto cuando no]

Quiero una "cosa" numérica que acepta la ecuación diferencial *y* la condición inicial, y regresa la solución $\mathbf{x}(t)$ para todas las $t$. (O más bien, para $t$ en un intervalo $[t_0, t_\text{final}]$

Eso existe --al menos, se puede tener una *aproximación* de la solución en unos tiempos dados.

*Discretizamos* el tiempo en pasos de tamaño $\delta t$

Necesitamos darle la siguiente información al método numérico:

- la condición inicial (un número, o un vector)
- la manera de calcular $\mathbf{f}$ como función de la posición actual, o sea una función
